```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("MagCurve.csv")

# Converter vírgulas para ponto decimal
df["MMF"] = df["MMF"].str.replace(",", ".").astype(float)
df["Fluxo"] = df["Fluxo"].str.replace(",", ".").astype(float)

# Número de espiras
N = 691.0994764397906

# Calcular a corrente
df["Corrente"] = df["MMF"] / N

# Definir o passo de tempo
dt = 1 / 3000  # segundos

# Criar vetor de tempo
df["Tempo"] = [i * dt for i in range(len(df))]

# Plotar Corrente x Tempo
plt.figure(figsize=(10, 5))
plt.plot(df["Tempo"], df["Corrente"], label="Corrente de Magnetização")
plt.xlabel("Tempo (s)")
plt.ylabel("Corrente (A)")
plt.title("Curva da Corrente de Magnetização x Tempo")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
